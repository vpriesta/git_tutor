```python
df_pd = pd.read_csv('IMDB-Movie-Data.csv')
df_pd
```

```python
def grade(x):
    if x > 8:
        r = "A"
    elif x > 5 and x <= 8:
        r = "B"
    else:
        r = "C"
    return r
```

```python
rt = []
for i in df_pd['Rating']:
    r = grade(i)
    rt.append(r)
    
# df_pd['Grade'] = rt
```

```python
df_pd['Grade'] = rt
df_pd
```

```python
df_re = pd.read_csv('Latihan Cleansing.csv')
df_re
```

```python
def cleanse(x):
    low = x.lower()
    cl = re.sub(r'[^a-zA-Z0-9]',' ',low)
    return cl
```

```python
tw = []
for i in df_re['Text Tweet']:
    c = cleanse(i)
    tw.append(c)

df_re['Clean Tweet'] = tw
df_re
```
